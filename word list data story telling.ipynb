```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vocab_WG = pd.read_csv('wordbank_vocab_data (1).csv')
vocab_WG
```

```python
vocab_WG['age_in_months'] = vocab_WG['age']
vocab_WG.drop('age', axis = 1, inplace = True)
vocab_WG
```

```python
_ = plt.scatter(vocab_WG['age_in_months'], vocab_WG['vocab'], alpha=0.5)
_ = plt.show()
```

```python
vocab_WG = vocab_WG[vocab_WG['age_in_months']<=18]
vocab_WG = vocab_WG[vocab_WG['age_in_months']>= 5]
```

```python
_ = plt.scatter(vocab_WG['age_in_months'], vocab_WG['vocab'], alpha=0.5)
_ = plt.show()
```

```python
median_list = []
for i in range(9,19):
    med_vocab = vocab_WG[vocab_WG['age_in_months'] == i]
    median_list.append(med_vocab['vocab'].median())
```

```python
print(median_list)
```

```python
x = list(range(9,19))
```

```python
_ = plt.plot(x, median_list)
_ = plt.show()
```

```python
std_list = []
for i in range(9,19):
    std_vocab = vocab_WG[vocab_WG['age_in_months'] == i]
    std_list.append(std_vocab['vocab'].std())
```

```python
_ = plt.plot(x, std_list)
_ = plt.show()
```

```python
print(std_list)
```

```python
zero_count = []
for i in range(9,19):
    zero_vocab = vocab_WG[vocab_WG['age_in_months'] == i]
    zero_count.append((zero_vocab['vocab'] == 0).sum()/len(zero_vocab))
```

```python
fifty_count = []
for i in range(9,19):
    fifty_vocab = vocab_WG[vocab_WG['age_in_months'] == i]
    fifty_count.append((fifty_vocab['vocab'] >= 50).sum()/len(fifty_vocab))
```
